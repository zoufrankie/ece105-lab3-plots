```python
# Generate simulated temperature readings from two sensors.
# Sensor A: mean 25 C, std 3 C, 200 readings.
# Sensor B: mean 27 C, std 4.5 C, 200 readings.
# Also generate 200 timestamps uniformly from 0 to 10 seconds.
# Use np.random.default_rng with a seed = last 4 digits of your Drexel ID.
import numpy as np
import matplotlib.pyplot as plt

# Set the seed for reproducibility
# Replace with the last 4 digits of your Drexel ID
rng = np.random.default_rng(seed=1234)

sensor_a = rng.normal(loc=25.0, scale=3.0, size=200)
sensor_b = rng.normal(loc=27.0, scale=4.5, size=200)
timestamps = rng.uniform(low=0.0, high=10.0, size=200)
print(sensor_a.shape)
```

```python
# Scatter plot of sensor data (assumes sensor_a, sensor_b, timestamps defined)
import numpy as np
import matplotlib.pyplot as plt

print('sensor_a shape:', sensor_a.shape)

# sort by time for clearer plotting
order = np.argsort(timestamps)
t = timestamps[order]
a = sensor_a[order]
b = sensor_b[order]

plt.figure(figsize=(8,4))
plt.scatter(t, a, s=30, alpha=0.7, label='Sensor A', color='C0')
plt.scatter(t, b, s=30, alpha=0.7, label='Sensor B', color='C1')
plt.title('Temperature Sensor Readings Over Time')
plt.xlabel('Time (s)')
plt.ylabel('Temperature (°C)')
plt.legend()
plt.tight_layout()
plt.show()

print(f"sensor_a: mean={a.mean():.2f}, std={a.std():.2f}")
print(f"sensor_b: mean={b.mean():.2f}, std={b.std():.2f}")
```

```python
# Overlaid histogram of Sensor A and Sensor B temperature distributions.
# Use 30 bins, alpha=0.5 for transparency so both distributions are visible.
# Add vertical dashed lines at each sensor's mean.
# Include a legend labeling each sensor.

plt.figure(figsize=(8,4))
plt.hist(sensor_a, bins=30, alpha=0.5, label='Sensor A', color='C0')
plt.hist(sensor_b, bins=30, alpha=0.5, label='Sensor B', color='C1')
plt.axvline(sensor_a.mean(), color='C0', linestyle='--', linewidth=2)
plt.axvline(sensor_b.mean(), color='C1', linestyle='--', linewidth=2)
plt.title('Temperature Sensor Readings Over Time')
plt.xlabel('Temperature (°C)')
plt.ylabel('Frequency')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Side-by-side box plot comparing Sensor A and Sensor B distributions.
# Label x-axis with sensor names, y-axis with "Temperature (deg C)".
# Add a horizontal dashed line at the overall mean of both sensors combined.

plt.figure(figsize=(8,4))
plt.boxplot([sensor_a, sensor_b], labels=['Sensor A', 'Sensor B'])
overall_mean = np.concatenate([sensor_a, sensor_b]).mean()
plt.title('Temperature Sensor Readings Over Time')
plt.axhline(overall_mean, color='red', linestyle='--', linewidth=2, label=f'Overall Mean: {overall_mean:.2f}°C')
plt.ylabel('Temperature (deg C)')
plt.legend()
plt.tight_layout()
plt.show()
```
